```python
# Jose Antonio Ceja Diaz
# 3/12/24
# pograma que cuente con un menu donde genere un dicionario, guarde los datos en una lista de diccionarios
# ademas de buscar, eliminar y ordenar el registro
# CDJA_ACT6.py



import random
import pandas as pd
import pprint
import json

# menu
def menu():
    claves = ["ID","APPAT","APMAT","NOMBRE","EDAD","SEXO","PUESTO"]
    lista_reg = []
    dicc_reg = {}
    op = 1
    band = 0
    band_ord = 0
    msg()
    while(op!=0):
        try:
            op = int(input("Ingresa una opcion: "))
        except:
            print("Favor de ingresar opcion valida")

        if(op == 1):
            for i in range(10):
                dicc_reg = crea_dicc(claves)
                lista_reg.append(dicc_reg)
            print("Registro generado exitosamente")
            band = 1
        elif(op==2):
            lista_reg.append(crea_reg_manu(claves))
            band == 1
        elif(op == 3):
            if(band == 1):
                try:
                    op2 = int(input("Selecciona como deseas imprimirlo: \n 1. pandas 2. yaml 3. pprint "))
                except:
                    print("Ingresa una opcion valida")
                if(op2 == 1):
                    salida=pd.DataFrame(lista_reg)
                    salida = salida.sort_values(by='ID')
                    print (salida)
                elif(op2 == 3):
                    pprint.pprint(lista_reg)
                elif(op2 == 2):
                    print(json.dumps(lista_reg, sort_keys=False, indent=2))
            else:
                print("No hay diccionario generado")
        elif(op == 4):
          if(band == 1):
              busc_id(lista_reg)
          else:
              print("No se han generado registros")
        elif(op == 5):
          if(band == 1 and band_ord != 1):
              ordena_reg(lista_reg)
              band_ord = 1
          else:
              print("No hay registro generado")
        elif(op == 6):
            if(band == 1):
              elimina_reg(lista_reg)
            else:
              print("No hay registro generado")
        elif(op == 7):
            if(band == 1):
              limpia_reg(lista_reg)
              print("lista de registro borrados correctamente")
              band = 0
            else:
              print("No hay registro generado")
        elif(op == 0):
            print("saliste del programa")


def msg():
    print("MENU")
    print("1. Crea diccionario (auto) ")
    print("2. Crea diccionario (manual) ")
    print("3. Imprime diccionario")
    print("4. Buscar(ID)")
    print("5. Ordenar")
    print("6. Eliminar ID")
    print("7. Eliminar toda la lista(ID)")
    print("0. Salir")

# funcion para generar valores
def gen_datos():
    nomh = ["luis","jose","alfredo"]
    nomm = ["grecia","maria","isabel"]
    appe = ["martinez", "Velazques","castilla"]
    puesto = ["Gerente","Empleado","Produccion","Contador","Administrativo","Reclutador"]

    lista=[]
    lista.append(random.randint(1,99999) + 300000)
    lista.append(random.choice(appe))
    lista.append(random.choice(appe))
    sex=random.randint(1,2)
    if (sex == 1):
        lista.append(random.choice(nomh))
    else:
        lista.append(random.choice(nomm))
    lista.append(random.randint(10,90))
    if (sex == 1):
        lista.append("HOMBRE")
    else:
        lista.append("MUJER")
    lista.append(random.choice(puesto))
    return lista

# funcion para crear diccionario
def crea_dicc(claves):
    valores = gen_datos()
    dicc_reg = dict(zip(claves,valores))
    return dicc_reg

def crea_reg_manu(claves):
    dict_reg ={"ID":"","APPAT":"","APMAT":"","NOMBRE":"","EDAD":"","SEXO":"","PUESTO":""}
    dict_reg["ID"] = int(input("Ingresa ID: "))
    dict_reg["APPAT"] = input("Ingresa apellido paterno: ")
    dict_reg["APMAT"] = input("Ingresa apellido materno: ")
    dict_reg["NOMBRE"] = input("Ingresa tu nombre: ")
    dict_reg["EDAD"] = int(input("Ingresa tu edad: "))
    dict_reg["SEXO"] = input("Ingresa tu sexo: ")
    dict_reg["PUESTO"] = input("Ingresa tu puesto de trabajo: ")
    print("Registro creado")
    return dict_reg

# funcion para imprimir lista
def impre_dicc(lista_reg):
    for i in range(10):
      print(lista_reg)

# Funcion para bucar id
def busc_id(lista_reg):
  try:
    id = int(input("que ID deseas buscar: "))
  except:
    print("Ingresa una matricula valida")
  for d in lista_reg:
    if d["ID"] == id:
      busq = d
      print(f"el registro de matricula: {id} es: {busq}")

# Funcion para ordenar registro
def ordena_reg(lista_reg):
    lista_reg.sort(key=lambda d:d["ID"])
    print("Registro ordenado correctamente")

# Elminar registro
def elimina_reg(lista_reg):
    try:
      id = int(input("que ID deseas eliminar: "))
    except:
      print("Ingresa una matricula valida")
    for d in lista_reg:
      if d["ID"] == id:
        lista_reg.remove(d)
        print(f"La matricula: {id} se elimino del registro")

def limpia_reg(lista_reg):
    lista_reg.clear()




menu()
```
